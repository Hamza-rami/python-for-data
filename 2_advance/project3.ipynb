what is pivot_table() function :

In Pandas, a pivot table is a data summarization tool that allows you to reorganize and aggregate data from a DataFrame to gain insights and analyze trends. It is conceptually similar to pivot tables found in spreadsheet software like Microsoft Excel. 

```python
# from datasets import load_dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# import data
df = pd.read_csv("/goinfre/hrami/python_for_data/data_jobs.csv")

# data cleanup
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
top_countries = df['job_country'].value_counts().head(6).index
jb_country_sal = df.pivot_table(values='salary_year_avg', index='job_country',columns='job_title_short' ,aggfunc='median')
top_countries_job = jb_country_sal.loc[top_countries]
job_title = ['Data Analyst'	,'Data Engineer' ,'Data Scientist']
top_countries_job = top_countries_job[job_title]
top_countries_job.plot(kind='bar')
plt.ylabel("median salary (USD)")
plt.title("median salary by country and job title")
plt.xticks(rotation=45, ha='right')
plt.show()
```
