matplotlib::pie plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ast

# import data
df = pd.read_csv("/goinfre/hrami/python/python_for_data/data_jobs.csv")

# data cleanup
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

```

```python
fig, ax = plt.subplots(1, 3)
dict = {'job_work_from_home':'Work from Home Status', 'job_no_degree_mention': 'No Degree Mention', 'job_health_insurance':'Health Insurance'}

for i, (colum, title) in enumerate(dict.items()):
    df[colum].value_counts().plot(
        kind='pie', 
        startangle=45, 
        autopct='%1.2f%%', 
        title=title, 
        ax=ax[i]
    )

fig.tight_layout()
fig.set_size_inches(15, 8)
```
