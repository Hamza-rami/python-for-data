```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# import data
df = pd.read_csv("/goinfre/hrami/python_for_data/data_jobs.csv")

# data cleanup
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
filter_df =  df[pd.notna(df['salary_year_avg'])].copy()
```

```python
def calc(salary):
    return salary * 1.02

filter_df['sal'] = filter_df['salary_year_avg'].apply(calc)
```
