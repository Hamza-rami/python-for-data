```python
# from datasets import load_dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# import data
df = pd.read_csv("/goinfre/hrami/python_for_data/data_jobs.csv")

# data cleanup
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
usa_job = df[df['job_country'] == 'United States'].copy()
usa_job
```

```python
usa_job['job_posted_monthy'] = usa_job['job_posted_date'].dt.strftime('%B')
usa_job
```

```python
pivot_usa = usa_job.pivot_table(index='job_posted_monthy', columns='job_title_short', aggfunc='size')
```

```python
pivot_usa.reset_index(inplace=True)
pivot_usa['month_nu'] = pd.to_datetime(pivot_usa['job_posted_monthy'], format='%B').dt.month
```

```python
pivot_usa.sort_values('month_nu', inplace=True)
pivot_usa.set_index('job_posted_monthy', inplace=True)
pivot_usa.drop(columns='month_nu', inplace=True)
```

```python
top_3 = df['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
top_3
```

```python
pivot_usa[top_3].plot(kind='line')
```
